## Convolutions

Illustrates the effect of applying convolutions.  A figure from chapter 13.

```julia
using Distributions
using PyPlot

xmin, xmax = -4, 3

function Q(s)
    if -1 < s < 0
        return 1 + s
    elseif 0 <= s < 1
        return 1 - s
    else
        return 0
    end
end
        
grid = linspace(xmin, xmax, 160)

fig, ax = subplots()

ax[:set_ylim](0, 1.1)
#ax[:set_yticks]((0.5, 1.0))
#ax[:set_xticks]((xmin, xmax))

ax[:plot](grid, [Q(g) for g in (grid)], "k-", lw=2, alpha=0.6, label=L"$Q$")

for h in (0.2, 0.5, 1.0)
    Kd = Normal(0, h)
    K(x) = pdf(Kd, x)
    function KstarQ(t)
        val, error = quadgk(s -> K(t - s) * Q(s), xmin, xmax)
        return val
    end
    lb = "\$K_h \\star Q\$ with \$h=$h\$"
    ax[:plot](grid, KstarQ(grid), lw=2.5, label=lb, alpha=0.5)
end

ax[:legend](frameon=0, loc="upper left", fontsize=15)
```
