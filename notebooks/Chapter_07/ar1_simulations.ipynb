## Generate AR(1) processes using R

This R code replicates the figure showing AR(1) simulations in chapter 7

```r
# Generates an AR(1) time series starting from init
ar1ts <- function(init, n, b, a) {
    x <- numeric(n)
    x[1] <- init
    w <- rnorm(n-1)
    for (t in 1:(n-1)) {
        x[t+1] <- b + a * x[t] + w[t]
    }
    return(x)
}

avec <- c(0.1, -0.1, 0.9, -0.9, 1.1, -1.1)
N <- 200


par(mfrow=c(3,2))  # Arrangement of figures
for (a in avec) {
    plot(ar1ts(0, N, 1, a), type="l",
        xlab=paste("a = ", a), ylab="")
}

dev.off()
```
