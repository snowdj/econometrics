## Metropolis Hastings random walk algorithm

This notebook gives a simple implementation of the Metropolis Hastings algorithm with random walk proposal.  It includes a test with a particular beta density.  Try plugging in different parameters or different densities.  The histogram should match whatever density you put in.

The code takes a while to start up because of the PyPlot library but will run fast after that.

```julia
using PyPlot
using Distributions

function rw_metropolis(pi_density, T, init=0, sigma=1)
    xvec = Array(Float64, T)
    xvec[1] = init
    for t in 1:(T - 1)
        x = xvec[t]
        y = x + sigma * randn()
        alpha = pi_density(x) > 0 ? pi_density(y) / pi_density(x) : 1
        xvec[t+1] = rand() < alpha  ? y : x
    end
    return xvec
end

a, b = 5, 5
beta_distribution = Beta(a, b)
beta_density(x) = pdf(beta_distribution, x)

beta_draws = rw_metropolis(beta_density, 10^6, 0.1)

fig, ax = subplots()
ax[:hist](beta_draws, bins=40, normed=1, alpha=0.4)
xgrid = linspace(0, 1, 140)
lb = "beta\$($a, $b)\$"
ax[:plot](xgrid, beta_density(xgrid), "k-", lw=2, alpha=0.6, label=lb)
ax[:legend](frameon=0)
```
